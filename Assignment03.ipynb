```python
List = [1, 2, 3, 2, 4, 5, 1]
seen = []
duplicates = []

for item in List:
    if item in seen:
        if item not in duplicates:
            duplicates.append(item)
    else:
        seen.append(item)

print(duplicates) # Output: [2, 1]
```
