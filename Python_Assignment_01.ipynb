#Builiding a Calculator Using Conditionals

```python
# creating 3 variables and getting input from users 2 variables for numbers and 3rd for operator

Num1 = int(input("Enter First Number: "))
Num2 = int(input("Enter Second Number: "))
Operator = input("Enter Operator: ")

if Operator == "+":
  print(Num1 + Num2)
elif Operator == "-":
  print(Num1 - Num2)
elif Operator == "*":
  print(Num1 * Num2)
elif Operator == "/": #float division
  if Num2 == 0:
    print("Invalid Division")
  else:
    print(Num1 / Num2)
else:
  print("Invalid Operator")
```

```python
#Question Number - 07
# Marksheet App
Eng = int(input("Enter English Marks:"))
Isl = int(input("Enter Isl Marks:"))
Math = int(input("Enter Maths Marks:"))
Urdu = int(input("Enter Urdu Marks:"))
Comp = int(input("Enter Comp Marks:"))
obtain = Eng + Isl + Math + Urdu + Comp
print("Your Obtain Marks is : ",obtain)
perc = obtain / 500 * 100
print("Your Percentage is : ", perc)
if perc >=80 and perc <=100:
    print("A-One Grade")
elif perc >= 70 and perc < 80:
    print("A Grade")
elif perc >= 60 and perc < 70:
    print("B Grade")
elif perc >= 50 and perc < 60:
    print("C Grade")
elif perc >= 40 and perc < 50:
    print("D Grade")
elif perc >= 33 and perc < 40:
    print("E Grade")
else:
    print("Fail")
```
